```python
import pandas as pd

# Load data
df_2024 = pd.read_csv('/content/drive/MyDrive/colab/data.csv')
```

```python
df_2024
```

```python
# Summary statistics
summary_stats = df_2024['Marks'].describe()
skewness = df_2024['Marks'].skew()
kurtosis = df_2024['Marks'].kurtosis()
```

```python
print("Summary Statistics:")
print(summary_stats)
print(f"Skewness: {skewness}")
print(f"Kurtosis: {kurtosis}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot histogram
plt.figure(figsize=(12, 6))
sns.histplot(df_2024['Marks'], kde=True)
plt.title('Distribution of Marks in NEET-UG-2024')
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.show()


```

```python
# Probability density function
plt.figure(figsize=(12, 6))
sns.kdeplot(df_2024['Marks'])
plt.title('Probability Density Function of Marks in NEET-UG-2024')
plt.xlabel('Marks')
plt.ylabel('Density')
plt.show()
```

```python
from scipy import stats

normality_test = stats.normaltest(df_2024['Marks'])
print(f"Normality Test - Statistic: {normality_test.statistic}, p-value: {normality_test.pvalue}")
```

The marks obtained in NEET-UG-2024 exhibit a right-skewed distribution with low kurtosis. This suggests a majority of students have lower scores with a few high scores, resulting in a distribution that is not normal.

```python
df_comparison = pd.read_csv('/content/drive/MyDrive/colab/data2023.csv')
```

```python
df_comparison
```

```python
pivot_df = df_comparison.pivot(index='score_bin', columns='year', values='frequency')
```

```python
pivot_df.plot(kind='bar', figsize=(12, 6))
plt.title('Comparison of Score Distributions between 2023 and 2024')
plt.xlabel('Score Bin')
plt.ylabel('Frequency')
plt.show()
```

The comparison between the 2023 and 2024 distributions suggests that the performance distribution of students remained relatively consistent between the two years. The similarity in distributions does not provide strong evidence of widespread cheating. However, the slightly higher frequencies in some score bins for 2024 could merit further investigation to understand if these are due to normal variance or other factors.
